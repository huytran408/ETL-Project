# Webscraping

```python
from bs4 import BeautifulSoup
import requests
import time
from splinter import Browser
```

```python
browser = Browser("chrome", executable_path="chromedriver", headless=False)

url = "https://www.kayak.com/flights/SFO-SGN/2019-12-20/2020-01-04?sort=bestflight_a"
browser.visit(url)
time.sleep(35)

html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
resultInner = soup.find_all("div",class_="resultInner")
```

```python
result = resultInner[3]
```

```python
result
```

```python
departure = result.findAll("div",{"class":"bottom"})
print(f'Depature Airline: {departure[0].text}')
print(f'Return Airline: {departure[3].text}')
```

```python
departure_time_container = result.findAll("span",{"class":"time-pair"})
print(f'Depature Time: {departure_time_container[0].text.strip()}')
```

```python
duration_container = result.findAll("div",{"class":"top"})
print(f'Duration: {duration_container[2].text}')
```

```python
price_container = result.findAll("span",{"class":"price option-text"})
price_container[0].text
```

```python
filename = "trips.csv"
f = open(filename, "w")

headers = "Departure_Airline, Return_Airline, Depature_time, Duration, Price\n"
f.write(headers)
```

```python
for container in resultInner:
    departure = container.findAll("div",{"class":"bottom"})
    departure_airline = departure[0].text
    return_airline = departure[3].text
    departure_time_container = result.findAll("span",{"class":"time-pair"})
    departure_time = departure_time_container[0].text.strip()
    departure_time = str.replace(departure_time,'\n','')    
    price_container = result.findAll("span",{"class":"price option-text"})
    price = price_container[0].text
    price = str.replace(price, '\n','')
    duration_container = result.findAll("div",{"class":"top"})
    duration_time = duration_container[2].text
    
    
    
    print("Departure Airline: " + departure_airline)
    print("Return Airline: " + return_airline)
    print("Departure Time: " + departure_time)
    print("Flight_Duration: " + duration_time)
    print("Price" + price)
    
    f.write(departure_airline +  "," + return_airline + "," + departure_time + "," + duration_time + "," + price + "\n" )
    
f.close()
    
    
```

```python
class Trips(Base):
    __tablename__ = 'trips'
    id = Column(Integer, primary_key=True, autoincrement=True)
    Departure = Column(String(20))
    Arrival = Column(Integer)
    Price = Column(Float)
```
